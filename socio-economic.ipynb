## Highlight insights related to demographic influences on income (age, gender, race, etc.).

```python
# Map the income brackets to numeric values
income_map = {'-50000': 0, ' 50000+.': 1}
df['Income_bracket_numeric'] = df['Income_bracket'].map(income_map)
```

```python
sns.lineplot(x="Age_of_the_individual", y="Income_bracket_numeric", data=df)
plt.xlabel('Age')
plt.ylabel('Income Bracket')
plt.title('Age By Income Braket')
plt.show()
```

- We can see that people in the age group of 40 to 60 are earning more, but their income drops dramatically as they get older.

```python
sns.lineplot(x="Age_of_the_individual", y="Income_bracket_numeric", hue="Sex_of_the_individual", data=df)
plt.xlabel('Age of the Individual')
plt.ylabel('Income Bracket')
plt.title('Income Level By Sex')
plt.show()
```

The graph provides information on the income bracket of individuals based on their age and sex. Here are the key takeaways:

- **Income Trends with Age**: Income tends to increase with age, peaking around 40-60 years old, and then gradually decreases.
- **Gender Differences**: There is a noticeable difference in income between males and females across all ages. Males generally have a higher income.

## Discuss the impact of socioeconomic factors such as education and employment type on income.

```python
# Adjust the figure size for better clarity
plt.figure(figsize=(13, 5))  

# Define a custom palette
palette = sns.color_palette("Set2")
# Create the bar plot with custom palette
sns.barplot(data=df, x='Highest_grade_attended', y='Income_bracket_numeric', hue='Sex_of_the_individual', palette=palette)

# Add title and labels with larger font size for better readability
plt.xlabel('Highest Grade Attended', fontsize=14)
plt.ylabel('Income Bracket (0: <=50K, 1: >50K)', fontsize=14)
plt.title('Income Level by Highest Grade Attended and Gender', fontsize=16)

# Rotate x-axis labels for better readability
plt.xticks(rotation="vertical", size=10)

# Add gridlines
plt.grid(True, linestyle='--', linewidth=0.5)

# Position the legend
plt.legend(title='Gender', loc='upper right')

# Show the plot
plt.show()
```

- Higher Education and Higher Income: Individuals who have attended higher levels of education, such as Bachelors, Masters, LLB, MD, etc., tend to have higher income levels compared to those with lower educational attainment.- 
Gender Consistency: This trend holds true for both males and females, highlighting the universal benefit of higher education in achieving greater financial earnings.

```python
# Adjust the figure size for better clarity
plt.figure(figsize=(13, 5))  


# Create the bar plot with custom palette
sns.barplot(data=df, x='Class_of_worker', y='Income_bracket_numeric', hue='Sex_of_the_individual', palette=palette)

# Add title and labels with larger font size for better readability
plt.xlabel('Class Of Work', fontsize=14)
plt.ylabel('Income Bracket (0: <=50K, 1: >50K)', fontsize=14)
plt.title('Class Of Work By Income level and Sex', fontsize=16)

# Rotate x-axis labels for better readability
plt.xticks(rotation="vertical", size=10)

# Add gridlines
plt.grid(True, linestyle='--', linewidth=0.5)

# Position the legend
plt.legend(title='Gender', loc='upper right')

# Show the plot
plt.show()
```

 #### Insights on Income by Employment Category

1. **Self-Employed (Incorporated)**: Individuals who are self-employed and have incorporated their businesses are earning good money. This may be because they are successfully running their own businesses.

2. **Federal Government Employees**: People working for the federal government also have relatively high incomes. The federal government seems to pay its employees better compared to local government positions.

3. **Unemployed Individuals**: As expected, those who are not working are not making any money. The government should pay attention to this issue and consider support measures.

4. **Lower Income Groups**: Certain working groups, like self-employed (not incorporated), private sector employees, local government employees, and state government employees, are earning less. There should be focus on improving the income levels for these categories.

## Illustrate economic factors like investment income or losses and their effects on overall income.


```python

# Calculate overall investment income
df['Investment_income'] = df['Capital_gains'] - df['Capital_losses'] + df['Dividend_value']

plt.figure(figsize=(13, 5))

# Create bar plot to illustrate investment income vs overall income
sns.barplot(data=df, x='Investment_income', y='Income_bracket', hue='Sex_of_the_individual' ,palette=palette)

# Add title and labels with larger font size for better readability
plt.xlabel('Investment Income', fontsize=14)
plt.ylabel('Income Bracket (0: <=50K, 1: >50K)', fontsize=14)
plt.title('Effect of Investment Income on Overall Income by Gender', fontsize=16)

# Add gridlines
plt.grid(True, linestyle='--', linewidth=0.5)

# Position the legend
plt.legend(title='Gender', loc='upper right')

# Show the plot
plt.show()
```

- People who earn less than 50,000 have significantly lower investment income compared to those who earn more than 50,000.
